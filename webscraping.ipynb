```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
r=requests.get(url)
```

```python
soup=BeautifulSoup(r.text,"lxml")
```

```python
price= soup.find_all("h4", class_="float-end price card-title pull-right")
prices=[]
```

```python
for i in price:
  p=i.text
  prices.append(p)
```

```python
print(prices)
```

```python
name= soup.find_all("a", class_= "title")
Names=[]
for i in name:
  d=i.text
  Names.append(d)
print(Names)
```

```python
import pandas as pd
dict={"Name": Names ,"price":prices}
```

```python
df=pd.DataFrame(dict)
```

```python
df
```

Data Cleaning


```python
df.drop_duplicates()
```

```python
df[["Brand","Series","Color"]]=df["Name"].str.split(" ",2,expand=True)
```

```python
df
```

```python
df["price"]=df["price"].str.lstrip("$")
```

```python
df["price"]=df["price"].astype(float)
```

EDA

```python
df1= df.groupby("Brand").price.mean().sort_values(ascending=False)
```

```python
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x=df1.index,y=df1.values)
plt.xlabel("Brand", color="Red")
plt.xticks(rotation=90)
plt.ylabel(" Average Price in Dollars", color="Red")
plt.title("Average Price of Each Brand")
```

```python
sns.countplot(x="Brand",data=df)
plt.xticks(rotation=90)
plt.title("Number of each brand of Computers available")
```
